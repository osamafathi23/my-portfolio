```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
df = pd.read_csv("TWO_CENTURIES_OF_UM_RACES.csv")
```

```python
df.head(10)
```

```python
df.shape
```

```python
df.dtypes
```

```python
#clean up the data
```

```python
#we want only usa races , 50k or 50 miles , 2020
```

```python
df[df['Event distance/length']=='50km']
```

```python
df[df['Event distance/length'].isin(['50km','50mi'])]
```

```python
df[(df['Event distance/length'].isin(['50km','50mi'])) & (df['Year of event']==2020)]
```

```python
df[(df['Event distance/length'].isin(['50km','50mi'])) & (df['Year of event']==2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
df2 = df[(df['Event distance/length'].isin(['50km','50mi'])) & (df['Year of event']==2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
df2.head(10)
```

```python
df2.shape
```

```python
#remove USA from Event name column
```

```python
df2['Event name'].str.split('(').str.get(0)
```

```python
df2['Event name'] = df2['Event name'].str.split('(').str.get(0)
```

```python
df2.head(10)
```

```python
df2['Athlete age category'] = df2['Year of event'] - df2['Athlete year of birth']
```

```python
df2.head(10)
```

```python
# remove h from Athlete performance
```

```python
df2['Athlete performance'] = df2['Athlete performance'].str.split('h').str.get(0)
```

```python
df2.head(10)
```

```python
# Clean up nulls
```

```python
df2.isna().sum()
```

```python
df2 = df2.drop(['Athlete club','Athlete year of birth'],axis=1)
```

```python
df2.head(10)
```

```python
df2.isna().sum()
```

```python
df2[df2['Athlete age category'].isna()==1]
```

```python
df2=df2.dropna()
```

```python
df2.isna().sum()
```

```python
df2.shape
```

```python
df2[df2.duplicated() == True]
```

```python
# no duplicates found
```

```python
#reset index
```

```python
df2.reset_index(drop=True)
```

```python
#fix Types
```

```python
df2.dtypes
```

```python
df2['Athlete age category'] = df2['Athlete age category'].astype(int)
```

```python
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float)
```

```python
df2.head()
```

```python
#rename cols
```

```python
df2 = df2.rename(columns= {
    'Year of event' : 'Year',
    'date':'race day',
    'Event distance/length':'Distance/Lenghth',
    'Athlete age category' : 'Athlete Age',
    'Event name' : 'race name'
})
```

```python
df2.head()
```

```python
#reorder Cols
```

```python
df3=df2[['Event dates','race name','Distance/Lenghth','Event number of finishers','Athlete Age','Athlete gender','Athlete country','Athlete average speed','Athlete performance','Athlete ID']]
```

```python
df3.head()
```

```python
sns.histplot(df3,x= 'Distance/Lenghth', hue = 'Athlete gender')
```
